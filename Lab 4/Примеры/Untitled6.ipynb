```python

import pandas as pd

s6 = pd.Series([1, 2, 3, 4, 5], ['a', 'b', 'c', 'd', 'e'])

print("Исходная Series:")
print(s6)

# ПРАВИЛЬНЫЙ доступ в pandas:
print("\n1. Доступ по ПОЗИЦИИ (числовой индекс) с помощью .iloc[]:")
print(f"s6.iloc[2] = {s6.iloc[2]}")   # третий элемент

print("\n2. Доступ по МЕТКЕ (индекс 'd') с помощью .loc[]:")
print(f"s6.loc['d'] = {s6.loc['d']}")

print("\n3. Доступ по МЕТКЕ (короткая форма):")
print(f"s6['d'] = {s6['d']}")
```

```python

import pandas as pd

s = pd.Series([10, 25, 8, 30, 15], index=['a', 'b', 'c', 'd', 'e'])

print("Исходная Series:")
print(s)

print("\nФильтрация значений в диапазоне от 10 до 30 (s[(s >= 10) & (s <= 30)]):")
filtered_s = s[(s >= 10) & (s <= 30)]
print(filtered_s)
```

```python

import pandas as pd

s = pd.Series([10, 25, 8, 30, 15], index=['a', 'b', 'c', 'd', 'e'])

print("Исходная Series:")
print(s)

print("\nВыбор элементов с индексами 'b' и 'd' (s[s.index.isin(['b', 'd'])]):")
filtered_s = s[s.index.isin(['b', 'd'])]
print(filtered_s)
```

```python

import pandas as pd
import numpy as np

s_with_nan = pd.Series([10, None, 8, 30, None], index=['a', 'b', 'c', 'd', 'e'])

print("Series с пропущенными значениями:")
print(s_with_nan)

print("\n.isnull() - обнаружение пропущенных значений:")
print(s_with_nan.isnull())

print("\n.notnull() - обнаружение непустых значений:")
print(s_with_nan.notnull())

print("\nФильтрация только непустых значений (s_with_nan[s_with_nan.notnull()]):")
filtered_s = s_with_nan[s_with_nan.notnull()]
print(filtered_s)
```

```python

import pandas as pd

s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

print("Исходная Series:")
print(s)

print("\nИзменение значения с индексом 'b' (s.loc['b'] = 25):")
s.loc['b'] = 25
print(s)

print("\nИзменение второго элемента (s.iloc[1] = 50):")
s.iloc[1] = 50
print(s)
```

```python

import pandas as pd

s = pd.Series([10, 20, 30, 40, 50, 60, 70], index=['a', 'b', 'c', 'd', 'e', 'f', 'g'])

print("Исходная Series:")
print(s)

print("\n.head(3) - первые 3 элемента:")
print(s.head(3))

print("\n.head() - первые 5 элементов (по умолчанию):")
print(s.head())

print("\n.tail(3) - последние 3 элемента:")
print(s.tail(3))

print("\n.tail() - последние 5 элементов (по умолчанию):")
print(s.tail())
```

```python

import pandas as pd

s = pd.Series([10, 20, 30, 40, 50], index=['a', 'b', 'c', 'd', 'e'])

print("Исходная Series:")
print(s)

print("\nИндексы (.index):")
print(s.index)

print("\nЗначения (.values):")
print(s.values)

print(f"\nПервый индекс (s.index[0]): {s.index[0]}")
print(f"Последний индекс (s.index[-1]): {s.index[-1]}")

print(f"\nПроверка 'b' в индексах ('b' in s.index): {'b' in s.index}")
print(f"Проверка 'z' в индексах ('z' in s.index): {'z' in s.index}")
```

```python

import pandas as pd

s = pd.Series([10, 20, 30, 40, 50])
print("Тип данных int64:", s.dtype)

s1 = pd.Series([1.5, 2.3, 3.7])
s2 = pd.Series(["apple", "banana", "cherry"])
s3 = pd.Series([True, False, True])

print("\ns1 (float64):", s1.dtype)
print("s2 (object):", s2.dtype)
print("s3 (bool):", s3.dtype)

print("\nПреобразование float в int (.astype(int)):")
s1_int = s1.astype(int)
print(s1_int)
print("Новый тип:", s1_int.dtype)
```
