```python
"""
Задание 1: Создание Series из списка

Задача:
Создайте Series из списка чисел [5, 15, 25, 35, 45] 
с индексами ['a', 'b', 'c', 'd', 'e']. 
Выведите его на экран и определите его тип данных.
"""

# Импортируем библиотеку pandas
import pandas as pd

# Создаём Series из списка с указанными индексами
s1 = pd.Series([5, 15, 25, 35, 45], index=['a', 'b', 'c', 'd', 'e'])

# Выводим Series на экран
print("Созданная Series:")
print(s1)

# Определяем и выводим тип данных
print(f"\nТип данных Series: {s1.dtype}")
```

```python
"""
Задание 2: Получение элемента Series

Задача:
Дан Series с индексами ['A', 'B', 'C', 'D', 'E'] 
и значениями [12, 24, 36, 48, 60]. 
Используйте .loc[] для получения элемента с индексом 'C' 
и .iloc[] для получения третьего элемента.
"""

# Импортируем библиотеку pandas
import pandas as pd

# Создаём Series
s2 = pd.Series([12, 24, 36, 48, 60], index=['A', 'B', 'C', 'D', 'E'])

# Выводим Series для наглядности
print("Исходная Series:")
print(s2)

# Получаем элемент с индексом 'C' с помощью .loc[]
element_loc = s2.loc['C']
print(f"\nЭлемент с индексом 'C' (s2.loc['C']): {element_loc}")

# Получаем третий элемент (индекс 2) с помощью .iloc[]
element_iloc = s2.iloc[2]
print(f"Третий элемент (s2.iloc[2]): {element_iloc}")
```

```python
"""
Задание 5: Определение типа данных Series

Задача:
Создайте Series из списка ['cat', 'dog', 'rabbit', 'parrot', 'fish']. 
Определите тип данных с помощью .dtype, 
затем преобразуйте его в category с помощью .astype().
"""

# Импортируем библиотеку pandas
import pandas as pd

# Создаём Series из списка строк
s5 = pd.Series(['cat', 'dog', 'rabbit', 'parrot', 'fish'])

# Выводим исходную Series
print("Исходная Series:")
print(s5)

# Определяем и выводим текущий тип данных
print(f"\nТекущий тип данных (dtype): {s5.dtype}")

# Преобразуем тип данных в 'category'
s5_category = s5.astype('category')

# Выводим результат преобразования
print("\nSeries после преобразования в category:")
print(s5_category)

# Выводим новый тип данных
print(f"\nНовый тип данных (dtype): {s5_category.dtype}")
```

```python
"""
Задание 6: Проверка пропущенных значений

Задача:
Создайте Series с данными [1.2, np.nan, 3.4, np.nan, 5.6, 6.8]. 
Напишите код, который проверяет, есть ли в Series пропущенные значения (NaN), 
и выведите индексы таких элементов.
"""

# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np

# Создаём Series с пропущенными значениями
s6 = pd.Series([1.2, np.nan, 3.4, np.nan, 5.6, 6.8])

# Выводим исходную Series
print("Исходная Series с пропущенными значениями:")
print(s6)

# Проверяем, есть ли пропущенные значения
has_nan = s6.isnull().any()
print(f"\nЕсть ли пропущенные значения в Series? {has_nan}")

# Находим индексы элементов, содержащих NaN
nan_indices = s6[s6.isnull()].index.tolist()
print(f"Индексы пропущенных значений (NaN): {nan_indices}")

# Выводим булевую маску для наглядности
print("\nБулевая маска (True - пропущенное значение):")
print(s6.isnull())
```

```python
"""
Задание 7: Заполнение пропущенных значений

Задача:
Используйте Series из предыдущего задания 
и замените все NaN на среднее значение всех непустых элементов. 
Выведите результат.
"""

# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np

# Создаём Series с пропущенными значениями
s7 = pd.Series([1.2, np.nan, 3.4, np.nan, 5.6, 6.8])

# Выводим исходную Series
print("Исходная Series с пропущенными значениями:")
print(s7)

# Вычисляем среднее значение непустых элементов
mean_value = s7.mean()
print(f"\nСреднее значение непустых элементов: {mean_value}")

# Заменяем NaN на среднее значение
s7_filled = s7.fillna(mean_value)

# Выводим результат
print("\nSeries после замены NaN на среднее значение:")
print(s7_filled)
```

```python
"""
Задание 8: Арифметические операции с Series

Задача:
Создайте два Series:
- s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
- s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])

Выполните сложение s1 + s2. 
Объясните, почему в результате появляются NaN, и замените их на 0.
"""

# Импортируем библиотеку pandas
import pandas as pd

# Создаём первый Series
s1 = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])

# Создаём второй Series
s2 = pd.Series([5, 15, 25, 35], index=['b', 'c', 'd', 'e'])

# Выводим оба Series
print("Series s1:")
print(s1)
print("\nSeries s2:")
print(s2)

# Выполняем сложение
s_sum = s1 + s2
print("\nРезультат сложения s1 + s2 (с NaN):")
print(s_sum)

# Объяснение появления NaN
print("\nПочему появляются NaN?")
print("Индексы 'a' из s1 и 'e' из s2 не имеют соответствующих пар в другой Series.")
print("Pandas не может выполнить операцию для этих индексов и вставляет NaN.")

# Заменяем NaN на 0 при сложении
s_sum_filled = s1.add(s2, fill_value=0)
print("\nРезультат сложения с заменой NaN на 0 (fill_value=0):")
print(s_sum_filled)
```

```python
"""
Задание 10: Основные статистические методы

Задача:
Создайте Series из 20 случайных чисел от 50 до 150 
(используйте np.random.randint). 
Найдите сумму, среднее, минимальное и максимальное значение. 
Выведите также стандартное отклонение.
"""

# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np

# Устанавливаем seed для воспроизводимости результатов
np.random.seed(42)

# Создаём Series из 20 случайных чисел от 50 до 150
s10 = pd.Series(np.random.randint(50, 151, size=20))

# Выводим исходную Series
print("Series из 20 случайных чисел от 50 до 150:")
print(s10)

# Вычисляем и выводим статистические показатели
print(f"\nСумма всех элементов: {s10.sum()}")
print(f"Среднее значение: {s10.mean():.2f}")
print(f"Минимальное значение: {s10.min()}")
print(f"Максимальное значение: {s10.max()}")
print(f"Стандартное отклонение: {s10.std():.2f}")

# Дополнительно выводим сводную статистику через .describe()
print("\nСводная статистика (.describe()):")
print(s10.describe())
```

```python
"""
Задание 12: Проверка уникальности индексов

Задача:
Создайте Series с индексами ['A', 'B', 'A', 'C', 'D', 'B'] 
и значениями [10, 20, 30, 40, 50, 60]. 
Проверьте, являются ли индексы уникальными. 
Если нет, сгруппируйте повторяющиеся индексы и сложите их значения.
"""

# Импортируем библиотеку pandas
import pandas as pd

# Создаём Series с повторяющимися индексами
s12 = pd.Series([10, 20, 30, 40, 50, 60], index=['A', 'B', 'A', 'C', 'D', 'B'])

# Выводим исходную Series
print("Исходная Series с повторяющимися индексами:")
print(s12)

# Проверяем, являются ли индексы уникальными
print(f"\nИндексы уникальны? {s12.index.is_unique}")

# Группируем повторяющиеся индексы и суммируем значения
s12_grouped = s12.groupby(s12.index).sum()

# Выводим результат группировки
print("\nРезультат группировки индексов (сумма значений для каждого индекса):")
print(s12_grouped)
```

```python
"""
Задание 15: Построение графика на основе Series

Задача:
Создайте Series, где индексами будут даты с 1 по 30 марта 2024 года, 
а значениями — случайные числа от 50 до 150. 
Постройте график значений с помощью matplotlib. 
Добавьте заголовок, подписи осей и сетку.
"""

# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Устанавливаем seed для воспроизводимости результатов
np.random.seed(42)

# Создаём диапазон дат с 1 по 30 марта 2024 года
dates = pd.date_range(start='2024-03-01', periods=30, freq='D')

# Создаём Series со случайными значениями от 50 до 150
s15 = pd.Series(np.random.randint(50, 151, size=30), index=dates)

# Выводим первые 5 элементов для ознакомления
print("Первые 5 элементов временного ряда:")
print(s15.head())

# Создаём график
plt.figure(figsize=(12, 6))
plt.plot(s15.index, s15.values, marker='o', linestyle='-', color='blue', markersize=4)

# Добавляем заголовок и подписи осей
plt.title('Временной ряд за март 2024 года', fontsize=14)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Значение', fontsize=12)

# Добавляем сетку
plt.grid(True, alpha=0.3)

# Поворачиваем подписи дат для лучшей читаемости
plt.xticks(rotation=45)

# Отображаем график
plt.tight_layout()
plt.savefig('grafico_zadanie15.png', dpi=150)
plt.show()

print("\nГрафик сохранён как 'grafico_zadanie15.png'")
```

```python
"""
Задание 17: Отслеживание выбросов в производственных данных

Задача:
Создайте Series, где индексами будут часы суток 
(pd.date_range('2024-11-15', periods=24, freq='H')), 
а значениями — случайное количество дефектных изделий (от 0 до 10). 
Преобразуйте индекс в DatetimeIndex, найдите выбросы (значения выше 8) 
и выделите их на графике.
"""

# Импортируем необходимые библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Устанавливаем seed для воспроизводимости результатов
np.random.seed(42)

# Создаём диапазон часов для 15 ноября 2024 года
hours = pd.date_range('2024-11-15', periods=24, freq='H')

# Создаём Series со случайным количеством дефектов от 0 до 10
defects = np.random.randint(0, 11, size=24)

# Добавляем искусственные выбросы (значения выше 8)
defects[3] = 9    # 03:00
defects[7] = 10   # 07:00
defects[14] = 9   # 14:00
defects[21] = 10  # 21:00

# Создаём Series
s17 = pd.Series(defects, index=hours)

# Выводим информацию о Series
print("Series почасовых данных о дефектах за 15.11.2024:")
print(s17)

# Находим выбросы (значения > 8)
outliers = s17[s17 > 8]
print(f"\nОбнаруженные выбросы (значения > 8):")
print(outliers)

# Создаём график
plt.figure(figsize=(14, 6))

# График всех значений
plt.plot(s17.index, s17.values, marker='o', linestyle='-', 
         color='blue', markersize=6, label='Количество дефектов')

# Выделяем выбросы красным цветом
plt.scatter(outliers.index, outliers.values, color='red', 
            s=120, zorder=5, label='Выбросы (>8)')

# Добавляем заголовок и подписи осей
plt.title('Почасовое количество дефектных изделий - 15 ноября 2024 года', fontsize=14)
plt.xlabel('Время (часы)', fontsize=12)
plt.ylabel('Количество дефектов', fontsize=12)

# Добавляем сетку
plt.grid(True, alpha=0.3)

# Поворачиваем подписи времени
plt.xticks(rotation=45)

# Добавляем легенду
plt.legend()

# Отображаем график
plt.tight_layout()
plt.savefig('grafico_zadanie17.png', dpi=150)
plt.show()

print("\nГрафик с выделенными выбросами сохранён как 'grafico_zadanie17.png'")
```
